# Importing Data & Libraries

```python
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
from config import *

games_df = pd.read_csv(f"{folder_path}/games.csv")
```

# Data Preparation of Games

```python
# handling NA values
games_df.isna().sum()
```

```python
# Toglie le righe in cui tutti i valori delle colonne elencate sono NaN
games_df = games_df.dropna(subset=['home_club_position','away_club_position','home_club_manager_name','away_club_manager_name','stadium','attendance','referee','home_club_formation','away_club_formation'], how='all')

games_df
```

```python
# Togliere le righe che non hanno nè squadra di casa nè squadra di trasferta
games_df = games_df.dropna(subset=['home_club_name','away_club_name'], how='all')
```

```python
games_df.isna().sum()
```

```python
# prendiamo i nomi delle squadre dal dataframe clubs, visto che abbiamo gli id delle squadre (in casa e in trasferta)

clubs_df = pd.read_csv(f"{folder_path}/clubs.csv")
clubs_df.columns
```

```python
games_df.columns
```

```python
# Aggiornamento di home_club_name
games_df = games_df.merge(clubs_df[['club_id', 'name']], how='left', left_on='home_club_id', right_on='club_id')
games_df['home_club_name'] = games_df['home_club_name'].fillna(games_df['name'])
games_df.drop(columns=['name', 'club_id'], inplace=True)

# Aggiornamento di away_club_name
games_df = games_df.merge(clubs_df[['club_id', 'name']], how='left', left_on='away_club_id', right_on='club_id')
games_df['away_club_name'] = games_df['away_club_name'].fillna(games_df['name'])
games_df.drop(columns=['name', 'club_id'], inplace=True)

games_df
```

```python
games_df.isna().sum()
```

```python
# le righe con il nome della squadra ancora NA sono quelle che non sono presenti nel csv clubs, quindi le rimuoviamo
games_df.dropna(subset=['home_club_name', 'away_club_name'])
```

```python
# gestione valori NA di position e attendance
games_df['home_club_position'].fillna(0, inplace=True)
games_df['away_club_position'].fillna(0, inplace=True)
games_df['attendance'].fillna(0, inplace=True)

# gestione valori NA delle formazioni
games_df['home_club_formation'].fillna('Unknown', inplace=True)
games_df['away_club_formation'].fillna('Unknown', inplace=True)
```

```python
games_df.dtypes
```

```python
# conversione date
games_df['date'] = pd.to_datetime(games_df['date']).dt.strftime('%d-%m-%Y')

# conversione position a int
games_df['home_club_position'] = games_df['home_club_position'].astype('int64')
games_df['away_club_position'] = games_df['away_club_position'].astype('int64')

# conversione attendance a int
games_df['attendance'] = games_df['attendance'].astype('int64')


games_df
```

```python
#making a dataframe subset and reordering columns
games_df = games_df [
    ['competition_id', 
     'season', 
     'round', 
     'date', 
     'home_club_id',
     'home_club_name',
     'home_club_goals',
     'home_club_position', 
     'home_club_manager_name',
     'home_club_formation', 
     'away_club_id',  
     'away_club_name',
     'away_club_goals',
     'away_club_position', 
     'away_club_manager_name', 
     'away_club_formation', 
     'stadium', 
     'attendance', 
     'referee', 
     #'url',
     'aggregate', 
     'competition_type']
]

games_df
```

```python
games_df = games_df.rename(columns={'aggregate':'score'})
```

```python
# check for home goal outliers
ax = games_df.boxplot(column='home_club_goals')
ax.set_title('Numbers of goals in Home Club')
plt.show()
```

```python
# check for away goal outliers
ax = games_df.boxplot(column='away_club_goals')
ax.set_title('Numbers of goals in Away Club')
plt.show()
```

```python
# attendance distribution
df = games_df.query('attendance > 0')
var = 'attendance'

mean_value = df[var].mean()
print(f"Media di {var}: {mean_value:.0f} spettatori")

sns.histplot(data=df, x=var, kde=True, bins=20)
plt.title(f'Distribuzione della variabile {var}')
plt.xlabel(f'{var}')
plt.ylabel('count')
plt.show()
```

```python
# goal segnati Casa

home_goal_counts = games_df['home_club_goals'].value_counts().sort_index()
total_games = len(games_df)
home_goal_percentages = (home_goal_counts / total_games) * 100


plt.figure(figsize=(10, 6))
ax = home_goal_percentages.plot(kind='bar', title="Numero di goal (Casa) in percentuale")

for i, percentage in enumerate(home_goal_percentages):
    ax.text(i, percentage + 0.5, f'{percentage:.2f}%', ha='center')

plt.xlabel('Goal Squadra in Casa')
plt.ylabel('Percentuale di partite (%)')
plt.xticks(rotation=0)
plt.ylim(0, 40)
plt.show()
```

```python
# goal segnati Trasferta

away_goal_counts = games_df['away_club_goals'].value_counts().sort_index()
total_games = len(games_df)
away_goal_percentages = (away_goal_counts / total_games) * 100


plt.figure(figsize=(10, 6))
ax = away_goal_percentages.plot(kind='bar', title="Numero di goal (Trasferta) in percentuale")

for i, percentage in enumerate(away_goal_percentages):
    ax.text(i, percentage + 0.5, f'{percentage:.2f}%', ha='center')

plt.xlabel('Goal Squadra in Trasferta')
plt.ylabel('Percentuale di partite (%)')
plt.xticks(rotation=0)
plt.ylim(0, 40)
plt.show()
```

```python
# adding a column name for the national leagues

league_names = {
    'BE1': 'Belgian Jupiler Pro League',
    'DK1': 'Danish Superliga',
    'ES1': 'LaLiga',
    'FR1': 'Ligue 1',
    'GB1': 'Premier League',
    'GR1': 'Greece Super League 1',
    'IT1': 'Serie A',
    'L1': 'Bundesliga',
    'NL1': 'Eredivisie',
    'PO1': 'Liga Portugal',
    'RU1': 'Russian Premier Liga',
    'SC1': 'Scottish Premiership',
    'TR1': 'Turkish Super League',
    'UKR1': 'Ukrainian Premier Liga'
}

competitions_df.loc[competitions_df['sub_type'] == 'first_tier', 'competition_name'] = competitions_df['competition_id'].map(league_names)
```

```python
# selecting only first tier leagues
competitions_df = competitions_df.dropna()
competitions_df
```

```python
# media di gol segnati a partita per campionato

games_df['total_goals'] = games_df['home_club_goals'] + games_df['away_club_goals']

# Raggruppa per campionato e stagione e calcola la media dei gol segnati a partita
avg_goals_per_game = games_df.groupby('competition_id')['total_goals'].mean().reset_index(name='avg_goals').sort_values(by='avg_goals', ascending=False)

# code by ChatGPT
mapping_name = competitions_df.set_index('competition_id')['competition_name'].to_dict()
avg_goals_per_game['competition_name'] = avg_goals_per_game['competition_id'].map(mapping_name)
avg_goals_per_game = avg_goals_per_game.dropna(subset=['competition_name'])


ax = sns.barplot(data=avg_goals_per_game, x='avg_goals', y='competition_name')

for i, avg_goals in enumerate(avg_goals_per_game['avg_goals']):
    ax.text(avg_goals + 0.05, i, f'{avg_goals:.2f}', color='black', ha="left", va='center')

plt.title('Media di goal a partita per ogni Campionato')
plt.xlabel('Media di goal a partita')
plt.ylabel('')
plt.xlim(1,4)
plt.show()
```
